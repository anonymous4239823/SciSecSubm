```python
# testing pcap upload function
from nfstream import NFStreamer
import os

def load_and_process_pcap(pcap_file_path):
    # Sprawdzenie, czy plik istnieje
    if not os.path.exists(pcap_file_path):
        print(f"Plik {pcap_file_path} nie istnieje.")
        return
    
    # Wczytywanie i przetwarzanie pliku pcapng za pomocą nfstream
    try:
        streamer = NFStreamer(source=pcap_file_path)
        for flow in streamer:
            print(flow)  # Możesz dostosować, co chcesz zrobić z przepływami danych
    except Exception as e:
        print(f"Wystąpił problem podczas przetwarzania pliku {pcap_file_path}: {e}")

# Ścieżka do pliku testing_small.pcapng, przy założeniu, że skrypt jest uruchamiany z głównego katalogu projektu
pcap_file_path = '../data/raw/testing_small.pcapng'

# Wywołanie funkcji
load_and_process_pcap(pcap_file_path)
```

```python
import pyshark

import nest_asyncio
nest_asyncio.apply()


def load_pcap_file(file_path):
    # Wczytanie pliku pcapng
    cap = pyshark.FileCapture(file_path)
    
    try:
        for packet in cap:
            # Wypisanie podstawowych informacji o pakiecie
            print(f"Packet Number: {packet.number}")
            print(f"Capture Length: {packet.captured_length}")
            print(f"Highest Layer: {packet.highest_layer}")
            print(f"Length: {packet.length}")
            print(f"Layers: {[layer.layer_name for layer in packet.layers]}")
            print("----------")
    except Exception as e:
        print(f"Wystąpił problem podczas wczytywania pliku {file_path}: {e}")
    finally:
        cap.close()

# Ścieżka do twojego pliku pcapng
file_path = '../data/raw/testing_small.pcapng'

# Wywołanie funkcji
load_pcap_file(file_path)
```
